```python
# Poverty rate based on CPS thresholds
def poverty_rate(
    income: np.ndarray, household_size: np.ndarray, number_children: np.ndarray, age: np.ndarray
) -> np.ndarray:

    pov_rate = np.zeros(income.shape[0]).astype(int)

    for key, value in enumerate(income):
        if household_size[key] == 1:
            if (age[key] < 65 and value < 13465) or (age[key] >= 65 and value < 12413):
                pov_rate[key] = 1
        elif household_size[key] == 2:
            if age[key] < 65 and (
                (number_children[key] == 0 and value < 17331) or (number_children[key] == 1 and value < 17839)
            ):
                pov_rate[key] = 1
            elif age[key] >= 65 and (
                (number_children[key] == 0 and value < 15644) or (number_children[key] == 1 and value < 17771)
            ):
                pov_rate[key] = 1
        elif household_size[key] == 3 and (
            (number_children[key] == 0 and value < 20244)
            or (number_children[key] == 1 and value < 20832)
            or (number_children[key] == 2 and value < 20852)
        ):
            pov_rate[key] = 1
        elif household_size[key] == 4 and (
            (number_children[key] == 0 and value < 26695)
            or (number_children[key] == 1 and value < 27131)
            or (number_children[key] == 2 and value < 26246)
            or (number_children[key] == 3 and value < 26338)
        ):
            pov_rate[key] = 1
        elif household_size[key] == 5 and (
            (number_children[key] == 0 and value < 32193)
            or (number_children[key] == 1 and value < 32661)
            or (number_children[key] == 2 and value < 31661)
            or (number_children[key] == 3 and value < 30887)
            or (number_children[key] == 4 and value < 30414)
        ):
            pov_rate[key] = 1
        elif household_size[key] == 6 and (
            (number_children[key] == 0 and value < 37027)
            or (number_children[key] == 1 and value < 37174)
            or (number_children[key] == 2 and value < 36408)
            or (number_children[key] == 3 and value < 35674)
            or (number_children[key] == 4 and value < 34582)
            or (number_children[key] == 5 and value < 33935)
        ):
            pov_rate[key] = 1
        elif household_size[key] == 7 and (
            (number_children[key] == 0 and value < 42605)
            or (number_children[key] == 1 and value < 42871)
            or (number_children[key] == 2 and value < 41954)
            or (number_children[key] == 3 and value < 41314)
            or (number_children[key] == 4 and value < 40124)
            or (number_children[key] == 5 and value < 38734)
            or (number_children[key] == 6 and value < 37210)
        ):
            pov_rate[key] = 1
        elif household_size[key] == 8 and (
            (number_children[key] == 0 and value < 47650)
            or (number_children[key] == 1 and value < 48071)
            or (number_children[key] == 2 and value < 47205)
            or (number_children[key] == 3 and value < 46447)
            or (number_children[key] == 4 and value < 45371)
            or (number_children[key] == 5 and value < 44006)
            or (number_children[key] == 6 and value < 42585)
            or (number_children[key] == 7 and value < 42224)
        ):
            pov_rate[key] = 1
        elif household_size[key] >= 9 and (
            (number_children[key] == 0 and value < 57319)
            or (number_children[key] == 1 and value < 57597)
            or (number_children[key] == 2 and value < 56831)
            or (number_children[key] == 3 and value < 56188)
            or (number_children[key] == 4 and value < 55132)
            or (number_children[key] == 5 and value < 53679)
            or (number_children[key] == 6 and value < 52366)
            or (number_children[key] == 7 and value < 52040)
            or (number_children[key] >= 8 and value < 50035)
        ):
            pov_rate[key] = 1

    return pov_rate
```

```python
acs2019_head_hh["poverty"] = poverty_rate(
    income=acs2019_head_hh["family_income"].to_numpy(),
    household_size=acs2019_head_hh["hh_size"].to_numpy(),
    number_children=acs2019_head_hh["nb_children"].to_numpy(),
    age=acs2019_head_hh["age"].to_numpy(),
)
```
